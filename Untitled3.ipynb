```python
perc = 50.0
min_count =  int(((100-perc)/100)*df.shape[0] + 1)
mod_df = df.dropna( axis=1, 
                thresh=min_count)
print("Modified Dataframe : ")
print(mod_df)
```

```python
mod_df.isnull().sum()
```

```python
mod_df.dropna(axis=0,inplace=True)
print(mod_df)
```

```python
mod_df.rename(columns = {'Embarked':'Embark'}, inplace = True)
print(mod_df)
```

```python
mod_df['Embark'] = mod_df['Embark'].replace(['C'],'0')
mod_df['Embark'] = mod_df['Embark'].replace(['Q'],'1')
mod_df['Embark'] = mod_df['Embark'].replace(['S'],'2')
print(mod_df)
```
